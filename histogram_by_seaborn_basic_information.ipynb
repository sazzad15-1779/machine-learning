# Histogram by seaborn 
If you have numeric type dataset and want to visualize in histogram then the seaborn histogram will help you. For this seaborn distplot function responsible to plot it.

## Import Libraries

```
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

## load data from github by seaborn

```
data =sns.load_dataset('tips')
data.head(2)
```

## from tips data set we get 3 numerical value total_bill,tip,size . now apply histogram

```
sns.distplot(data['size'])
```

```
sns.distplot(data['tip'])
```

```
sns.distplot(data['total_bill'])
```

**Syntax: sns.distplot(      
                                   a,
                                     bins=None,
                                     hist=True,
                                     kde=True,
                                     rug=False,
                                     fit=None,
                                     hist_kws=None,
                                     kde_kws=None,
                                     rug_kws=None,
                                     fit_kws=None,
                                     color=None,
                                     vertical=False,
                                     norm_hist=False,
                                     axlabel=None,
                                     label=None,
                                     ax=None,
                               )**

## bins

```
sns.set()
sns.distplot(data['total_bill'],bins=[0,5,10,15,20,25,30,35,40,45,50,55,60])
```

```
sns.set()
sns.distplot(data['total_bill'],bins = 40)
```

## hist , kde ,rug , hist_kws, kde_kws , rug_kws

```
sns.set()
sns.distplot(data['tip'],bins=20)
```

```
sns.set()
sns.distplot(data['tip'],bins=20,hist =False)  # histogram False
```

```
sns.set()
sns.distplot(data['tip'],bins=20,kde =False)  # kernal density estimator(line) =False 
```

```
sns.set()
sns.distplot(data['tip'],rug = True) # rug true
```

#### hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5, 'linestyle':'--', 'alpha':0.9}


```
flight = sns.load_dataset('flights')
flight.head()
```

```
sns.distplot(flight['passengers'],hist_kws={'color':'orange','edgecolor':'black','linewidth':4,'linestyle':':','alpha':.7})
```

#### kde_kws = {'color':'#8e00ce','linewidth':8, 'linestyle':'--', 'alpha':0.9}

```
sns.distplot(flight.passengers,kde_kws= {'color':'red','linewidth':3,'linestyle':'--','alpha':.5})
```

```
sns.distplot(flight.passengers,kde_kws= {'color':'b','linewidth':2,'linestyle':'--','alpha':.7},hist_kws = {'color':'orange','edgecolor':'black','linewidth':3,'linestyle':':','alpha':.5})
```

#### rug_kws = {'color':'#0426d0', 'edgecolor':'#00dbff',linewidth':5, 'linestyle':'--', 'alpha':0.9}

```
iris = sns.load_dataset('iris')
iris.head(2)
```

```
sns.distplot(iris.sepal_length,rug = True,rug_kws={'color':'r','linewidth':2,'linestyle':"--","alpha":.7})
```

#### 1. fit
#### 2. fit_kws = {'color':'#8e00ce', 'linewidth':12, 'linestyle':'--', 'alpha':0.4}



```
from scipy.stats import norm
sns.distplot(iris.petal_length,fit= norm)
```

```

sns.distplot(iris.petal_length,fit= norm,kde=False)
```

```

sns.distplot(iris.petal_length,fit= norm,fit_kws={'color':"r",'linewidth':4,'linestyle':'--','alpha':.5},kde_kws={'color':'b','linewidth':4,'linestyle':':','alpha':.7})
```

## color , axlabel,lable 

```
sns.distplot(iris.sepal_length,color='g')
```

```
sns.distplot(iris.sepal_length,color='g',axlabel='iris',label ='sepal length')
plt.legend()
```
